```python
#check for null values
data.isnull().sum()
```

```python
#create independent and target variables
X = data[['pos', 'age', 'bref_team_id', 'g', 'gs', 'mp', 'fg', 'fga',
       'fg.', 'x3p', 'x3pa', 'x3p.', 'x2p', 'x2pa', 'x2p.', 'efg.', 'ft',
       'fta', 'ft.', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf']]
y = data['pts']
```

```python
#Taking care of missing data
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X.iloc[:,8:28])
X.iloc[:,8:28] = imputer.transform(X.iloc[:, 8:28])
```

```python
#check for isnull
X.isnull().sum()
```

```python
#get all the non-numerical values
obj_cols = X.dtypes[data.dtypes == 'object'].index
obj_cols
```

```python
#create and append dummy variables
for cols in obj_cols:
    X = X.join(pd.get_dummies(data[cols], prefix=cols).iloc[:,:-1])
```

```python
X.head()
```

```python
#drop non-numerical columns 
X = X.drop(obj_cols, axis=1)
```

```python
X.head()
```

```python
#k-nearest neighbour
import numpy as np
from sklearn.cross_validation import train_test_split
```

```python
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=70)
```

```python
# loading library
from sklearn.neighbors import KNeighborsClassifier
```

```python
# instantiate learning model (k = 10)
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn
```

```python
# fitting the model
knn.fit(X_train, y_train)
```

```python
# predict the response
pred = knn.predict(X_test)
```

```python
import matplotlib.pyplot as plt
#plt.scatter(X_train, y_train, color='red')
plt.scatter(y_test, pred)
plt.title('Predicted vs actual score')
plt.xlabel('test scores')
plt.ylabel('predicted scores')
```
